```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
df = (pd.DataFrame(load_iris().data, columns=load_iris().feature_names)).iloc[:, [2,3]]
X = df.values
df.head()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    print(f'K={i}, Inertia={kmeans.inertia_}')
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
cluster_pred = kmeans.predict(X)
cluster_pred
```

```python
plt.scatter(X[cluster_pred==0,0], X[cluster_pred==0,1], s=100, c='red', label='Cluster 1')
plt.scatter(X[cluster_pred==1,0], X[cluster_pred==1,1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[cluster_pred==2,0], X[cluster_pred==2,1], s=100, c='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
```
